````python
# Since I was invited to use a chatbot and I didn't bother writing the code
# myself, I am pasting the chatbot's code below.
import numpy as np
from scipy.stats import norm

def bootstrap_sample(data, n_bootstrap):
    """ Generate bootstrap samples from the dataset. """
    n = len(data)
    bootstrap_means = np.zeros(n_bootstrap)
    for i in range(n_bootstrap):
        sample = np.random.choice(data, size=n, replace=True)
        bootstrap_means[i] = np.mean(sample)
    return bootstrap_means

def compute_confidence_interval(bootstrap_means, confidence_level=0.95):
    """ Compute the confidence interval from bootstrap samples. """
    lower_percentile = (1 - confidence_level) / 2 * 100
    upper_percentile = (1 + confidence_level) / 2 * 100
    lower_bound = np.percentile(bootstrap_means, lower_percentile)
    upper_bound = np.percentile(bootstrap_means, upper_percentile)
    return lower_bound, upper_bound

# Parameters
mu = 0  # mean of the normal distribution
sigma = 1  # standard deviation of the normal distribution
sample_size = 1000  # size of the sample
n_bootstrap = 10000  # number of bootstrap samples

# Generate a sample from a normal distribution
data = norm.rvs(size=sample_size, loc=mu, scale=sigma)

# Bootstrap the sample
bootstrap_means = bootstrap_sample(data, n_bootstrap)

# Compute the 95% confidence interval
confidence_level = 0.95
ci_lower, ci_upper = compute_confidence_interval(bootstrap_means, confidence_level)

print(f"95% Confidence Interval: [{ci_lower:.3f}, {ci_upper:.3f}]")
````

### What is the process of bootstrapping?
So, imagine you have a big bag of marbles with numbers on them, but instead of counting and checking every single marble, you only grab a handful to get a quick idea of what numbers might be in the entire bag. This handful of marbles is like your sample from a larger population.

Bootstrapping is like this: after you've taken your handful (your sample), you pretend that this handful is a mini version of the whole bag. You then start drawing marbles from this handful, noting down the number, and putting it back in the handful before drawing again. This way, you simulate a new handful (or sample), and you do this lots of times to see how much the numbers can vary each time. 

In technical terms, you randomly resample your original data (the handful of marbles) with replacement (putting the marbles back after drawing) to create many new samples. This helps in understanding the variability of your estimates (like averages or medians) from your original sample.

### What is the main purpose of bootstrapping?
The main goal of bootstrapping is to get a better understanding of the statistical properties (like average, standard deviation, etc.) of your original sample. Since you often donât have access to the whole population (all marbles in the bag), bootstrapping allows you to mimic the process of obtaining new samples, so you can estimate the uncertainty or accuracy of your statistical estimates.
